# **Preprocessing**

```python
import re
import ast
def convert_text(text):
    try:
        teks = ast.literal_eval(text)
        return " ".join([word.lower() for word in teks]) # Convert to lowercase here
    except (ValueError, SyntaxError):
        if isinstance(text, str):
          return " ".join([word.lower() for word in text.split()]) # Handle string cases
        return text

df_sample['content'] = df_sample['content'].apply(convert_text)
```

```python
def cleaning_text(text) :
    text = re.sub(r'#\w+', '', text) #remove hashtag
    text = re.sub(r'\b\w*\d\w*\b', '', text) #remove words with numbers in the middle
    text = re.sub(r'@\S+', '', text) #remove mentions (@username)
    text = re.sub(r'https?://\S+|www\.\S+', '', text) #remove URLs
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text) #remove special characters
    return text

df_sample['clean_content'] = df_sample['content'].apply(cleaning_text)
```

```python
nltk.download('punkt_tab')
df_sample['tokenized'] = df_sample['clean_content'].apply(word_tokenize)
```

```python
df_sample['tag'] = df_sample['tags'].apply(cleaning_text)
df_sample['tag'] = df_sample['tag'].apply(word_tokenize)
```

```python
df_sample = df_sample[df_sample['tag'].apply(lambda x: isinstance(x, list) and len(x)>0)]
df_sample.reset_index(drop=True, inplace=True)
df_sample.shape
```

```python
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('english')
df_sample['stopword'] = df_sample['tokenized'].apply(lambda x: [word for word in x if word.lower() not in stop_words])
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
type(df_sample['stopword'].iloc[0])
```

```python
lematizer = WordNetLemmatizer()
df_sample['lematized'] = df_sample['stopword'].apply(lambda x: [lematizer.lemmatize(word) for word in x]).apply(lambda x: " ".join(x))
```

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('english')
df_sample['stemming'] = df_sample['stopword'].apply(lambda x: [stemmer.stem(word) for word in x])
```

# **TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix_lem = tfidf_vectorizer.fit_transform(df_sample['lematized'])
tfidf_similarity_lem = cosine_similarity(tfidf_matrix_lem, tfidf_matrix_lem)
print(f"TF-IDF Matrix Shape: {tfidf_matrix_lem.shape}")
```

```python
tfidf_csvector_lem = pd.DataFrame(tfidf_similarity_lem, index=df_sample['name'], columns=df_sample['name'])
tfidf_csvector_lem
```

```python
tfidf_matrix_stem = tfidf_vectorizer.fit_transform(df_sample['stemming'].apply(lambda x: ' '.join(x)))
tfidf_similarity_stem = cosine_similarity(tfidf_matrix_stem, tfidf_matrix_stem)
print(f"TF-IDF Matrix Shape: {tfidf_matrix_stem.shape}")
tfidf_csvector_stem = pd.DataFrame(tfidf_similarity_stem, index=df_sample['name'], columns=df_sample['name'])
tfidf_csvector_stem
```

```python
def rec_tfidf_lem(game_name, top=10):
    if game_name not in tfidf_csvector_lem.index:
        print(f"Game '{game_name}' tidak ditemukan dalam TF-IDF matrix.")
        return pd.DataFrame()

    similarity_scores = tfidf_csvector_lem.loc[game_name]
    sorted_scores = similarity_scores.sort_values(ascending=False)
    sorted_scores = sorted_scores.drop(labels=game_name, errors='ignore')

    # Ambil top N game paling mirip
    top_game_names = sorted_scores.head(top).index
    top_scores = sorted_scores.head(top).values
    recommended = df_sample[df_sample['name'].isin(top_game_names)][['name', 'release_year', 'content']].copy()

    # Tambahkan kolom skor similarity berdasarkan urutan yang benar
    recommended['similarity_score'] = recommended['name'].apply(lambda x: similarity_scores[x])
    recommended = recommended.set_index('name').loc[top_game_names].reset_index()

    print(f"Top {top} recommendations using lemmatization for '{game_name}':")

    return recommended
```

```python
def rec_tfidf_stem(game_name, top=10):
    if game_name not in tfidf_csvector_stem.index:
        print(f"Game '{game_name}' tidak ditemukan dalam TF-IDF matrix.")
        return pd.DataFrame()

    similarity_scores = tfidf_csvector_stem.loc[game_name]
    sorted_scores = similarity_scores.sort_values(ascending=False)
    sorted_scores = sorted_scores.drop(labels=game_name, errors='ignore')

    # Ambil top N game paling mirip
    top_game_names = sorted_scores.head(top).index
    top_scores = sorted_scores.head(top).values
    recommended = df_sample[df_sample['name'].isin(top_game_names)][['name', 'release_year', 'content']].copy()

    # Tambahkan kolom skor similarity berdasarkan urutan yang benar
    recommended['similarity_score'] = recommended['name'].apply(lambda x: similarity_scores[x])
    recommended = recommended.set_index('name').loc[top_game_names].reset_index()

    print(f"Top {top} recommendations using stemming for '{game_name}':")

    return recommended
```

```python
import gensim
from gensim.models import Word2Vec
```

```python
w2v = Word2Vec(df_sample['lematized'], vector_size=100, window=5,
               min_count=1, sg=0, workers=4, epochs=50)

# Rata-rata vektor
def get_vector(teks):
    teks = [word for word in teks if word in w2v.wv.key_to_index]
    if len(teks)==0:
        return np.zeros(w2v.vector_size)
    return np.mean(w2v.wv[teks], axis=0)

teks_vector = np.array([get_vector(teks) for teks in df_sample['lematized']])
w2v_cosim = cosine_similarity(teks_vector, teks_vector)
```

```python
w2v_csvector = pd.DataFrame(w2v_cosim, index=df_sample['name'], columns=df_sample['name'])
w2v_csvector
```

```python
def rec_w2v(game_name, top=10):
    if game_name not in df_sample['name'].values:
        print(f"Game '{game_name}' tidak ditemukan dalam data.")
        return pd.DataFrame()
    
    cosim_score = w2v_csvector.loc[game_name].sort_values(ascending=False)
    # sorted_scores = cosim_score.drop(labels=game_name, errors='ignore')
    
    top_game_name = cosim_score.head(top).index
    recommended = df_sample[df_sample['name'].isin(top_game_name)][['name', 'release_year', 'detailed_description','tag']].copy()
    recommended['similarity_score'] = recommended['name'].apply(lambda x: cosim_score.get(x, np.nan))
    recommended = recommended.set_index('name').loc[top_game_name].reset_index()
    
    print(f"Top {top} recommendations for '{game_name}':")
    return recommended.reset_index(drop=True)
```
