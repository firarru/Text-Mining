# **Text Summarization**

```python
import nltk.data
from nltk import tokenize
import string
```

```python
# !wget https://raw.githubusercontent.com/firarru/Text-Mining/main/Dataset/Article.txt

with open('Article.txt', 'r', encoding='utf-8') as file:
    text = file.read().replace('A', '').replace('â€', '').replace('\xa0','')
text_sentence = tokenize.sent_tokenize(text)
sentence_count = len(text_sentence)
print(sentence_count)
```

```python
#Count word frequency of sentence
wordFrequencyAllSentence = {}
for sentence in text_sentence:
    wordFrequencyPerSentence = {}
    word_list = sentence.split()

    for word in word_list:
        wordFrequencyPerSentence[word] = word_list.count(word)
        wordFrequencyAllSentence[word] = wordFrequencyPerSentence
```

```python
wordFrequencyAllSentence
```

TF(word) = number of times term word appears in a sentence / total number of terms in the sentence

```python
from nltk import word_tokenize

TFwordAllSentences = {}

for sentence in text_sentence :
  TFwordPerSentences = {}
  word_list = word_tokenize(sentence.lower())
  word_list = [word.lower() for word in word_list if word.isalpha()]

  for word in word_list:
    TFwordPerSentences[word] = word_list.count(word) / len(word_list)
    TFwordAllSentences[sentence] = TFwordPerSentences
```

for each word compute how many sentences contain that word

```python
from nltk import word_tokenize
full_text = " ".join(text_sentence)
allWords = word_tokenize(full_text.lower())
allWords = [word.lower() for word in allWords if word.isalpha()]
allWords = set(allWords)

word_count = dict()

for word in allWords:
  count = 0
  for sentence in text_sentence:
    if(word in sentence.lower()):
      count += 1
    word_count[word] = count
  word_count
```

```python
import math
IDForwardAllSentences = {}

totalSentences = len(text_sentence)
for sentence in text_sentence :
  IDForwardPerSentence = {}

  word_list = word_tokenize(sentence.lower())
  word_list = [word.lower() for word in word_list if word.isalpha()]

  for word in word_list:
    if(word_count[word]==0):
      pass
    else:
      IDForwardPerSentence[word] = math.log10(totalSentences/word_count[word])
      IDForwardAllSentences[sentence] = IDForwardPerSentence

IDForwardAllSentences
```

TF-IDF

```python
TF_IDFwordAllSentences = TFwordAllSentences

for key1, value1 in TF_IDFwordAllSentences.items():
  for key2, value2 in value1.items():
    TF_IDFwordAllSentences[key1][key2] = TFwordAllSentences[key1][key2] * IDForwardAllSentences[key1][key2]
    break

TF_IDFwordAllSentences
```

Threshold compute average sentence weight

```python
weight_sentences = {}

for key1, value1 in TF_IDFwordAllSentences.items():
  x=0
  for key2, value2 in value1.items():
    x = x+TF_IDFwordAllSentences[key1][key2]
    weight_sentences[key1] = x/len(key1)

weight_sentences
```

```python
averageSentencewieght = sum(list(weight_sentences.values())) / len(weight_sentences)
print(averageSentencewieght)
```

generate summary : select a sencentece for summarizations if the weight of sentence exceeds the threshold

```python
for key, value in weight_sentences.items():
  if(value > averageSentencewieght):
    print(key)
```
