# **Text Summarization**

```python
import nltk.data
from nltk import tokenize
import string
```

```python
# !wget https://raw.githubusercontent.com/firarru/Text-Mining/main/Dataset/Article.txt

with open('Article.txt', 'r', encoding='utf-8') as file:
    text = file.read().replace('A', '').replace('â€', '').replace('\xa0','')
text_sentence = tokenize.sent_tokenize(text)
sentence_count = len(text_sentence)
print(sentence_count)
```

```python
#Count word frequency of sentence
wordFrequencyAllSentence = {}
for sentence in text_sentence:
    wordFrequencyPerSentence = {}
    word_list = sentence.split()

    for word in word_list:
        wordFrequencyPerSentence[word] = word_list.count(word)
        wordFrequencyAllSentence[word] = wordFrequencyPerSentence
```

```python
wordFrequencyAllSentence
```

TF(word) = number of times term word appears in a sentence / total number of terms in the sentence

```python
from nltk import word_tokenize

TFwordAllSentences = {}

for sentence in text_sentence :
  TFwordPerSentences = {}
  word_list = word_tokenize(sentence.lower())
  word_list = [word.lower() for word in word_list if word.isalpha()]

  for word in word_list:
    TFwordPerSentences[word] = word_list.count(word) / len(word_list)
    TFwordAllSentences[sentence] = TFwordPerSentences
```
